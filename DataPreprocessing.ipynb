```python
import json
import pandas as pd
import numpy as np
import spacy
```

```python
# Reading json in
with open('./data/positive_tweet_data.jsonl', 'r') as json_file:
    json_list = list(json_file)
for json_str in json_list:
    result = json.loads(json_str)
    print("result: {}".format(result))
    print(isinstance(result, dict))
```

```python
len(result.keys())
```

```python
# Showing one result would be like this
result['1250216941891002369']
```

```python
# Putting the id, text, tweet time, coordinates, country and area into dataframes
extracted_list = []
for key, tweet in result.items():
    if tweet["place"]:
        extracted_list.append([tweet['id_str'], tweet['full_text'], tweet['created_at'],
                  tweet['place']['bounding_box']['coordinates'], tweet['place']['country'], tweet['place']['full_name']])
    else:
        extracted_list.append([tweet['id_str'], tweet['full_text'], tweet['created_at'],
                  None, None, None])
df = pd.DataFrame(extracted_list, columns = ['key', 'orginal_tweet_content','created_at','coordinates','country','area']) 
```

```python
# Showing how many tweets have location information
df[df['coordinates'].notnull()]
```

```python
# Spacy text preprocessing

language_model = spacy.load("en_core_web_sm")

df['doc'] = df['orginal_tweet_content'].apply(language_model)
```

```python
# What do we need? 
print(df.iloc[1]['doc'].ents)
print([t.dep_ for t in df.iloc[1]['doc']])
print([t.pos_ for t in df.iloc[1]['doc']])
print(df.iloc[1]['orginal_tweet_content'])
```
